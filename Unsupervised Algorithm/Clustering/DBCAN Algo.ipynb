```python
import numpy as np
import pandas as pd 

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
```

```python
X,y = make_moons(n_samples=250, noise=0.05)
```

```python
data = {"data1" : X [:,0] , "data2" : X [:,1]} 
df = pd.DataFrame(data)
```

```python
sns.pairplot(data = df )
plt.show()
```

```python
# epsilon = Maximum distance between two samples for one to be considered as in the neighborhood of the other
# min-samples = The number of samples in a neighborhood for a point to be considered as a core point

dbscan = DBSCAN(eps=0.2, min_samples=5)
df["predict"] = dbscan.fit_predict(df)
```

```python
df.head(8)
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x='data1', y='data2', hue='predict', palette='rainbow', s=100)
plt.title('DBSCAN Clusters of make_moons Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
